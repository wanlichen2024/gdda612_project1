```python
# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# define an empty list to store useful product infomation
all_items = []

# locate the target area of information
products = soup.select('div.product-inner')

# extra product name and price
for product in products:
    # product name
    name_tag = product.select_one('h2.woo-loop-product__title a')
    if name_tag:
        name = name_tag.text.strip()
    else:
        name = "N/A"

    # product price
    price_tag = product.select_one('span.price bdi')
    if price_tag:
        price = price_tag.text.strip()
    else:
        price = "N/A"

    # store all extracted information to the dictionary
    item = {
        'product_name': name,
        'product_price': price
    }

    # add dictionary to the list
    all_items.append(item)

# convert all information to csv format
with open('all_products.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=['product_name', 'product_price'])
    writer.writeheader()
    writer.writerows(all_items)

print("All information has been saved to all_products.csv")
```
