```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/all_products_lm.csv')
```

```python
df1.head()
```

```python
df1.shape
```

```python
#remove irrelevant data
# extract only contain iPhone rows
iphone_df=df1[df1['product_name'].str.contains('iPhone')]
iphone_df
```

```python
df1=iphone_df
```

```python
df1 = df1.drop_duplicates(keep='first')
print("\n keep the first record and remove the other duplicated value:")
print(df1)
```

```python
#check missing value
missing_value1=df1.isna().sum()
missing_value1
```

```python
missing_value_row1=df1[df1.isna().any(axis=1)]
missing_value_row1
```

```python
#Using mode to fill missing value
df1['product_rating'].fillna(df1['product_rating'].mode()[0],inplace=True)
```

```python
missing_value1=df1.isna().sum()
missing_value1
```

```python
# identify the pattern of column 'product_name' for spliting
pattern1 = r'(?P<brand>\w+)\s(?P<model>[\w\s]+?)\s(?P<storage>\d+(?:GB|TB))(?:\s-\s|\s)?(?P<colour>[\w\s]+)?'

# split column
df1[['brand', 'model', 'storage', 'colour']] = df1['product_name'].str.extract(pattern1)

print(df1)
```

```python
df1=pd.DataFrame(df1)
```

```python
# convert_storage_to_int for df1
def convert_storage_to_int(storage):
    return int(storage.replace('GB', '').replace('TB', '000')) if 'TB' in storage else int(storage.replace('GB', ''))
```

```python
df1['storage_int'] = df1['storage'].apply(convert_storage_to_int)
```

```python
df1_encode = df1[['model', 'storage_int', 'colour']]
print(df1_encode)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# create a new empty list
label_encoders = {}

# encode columns
for column in df1_encode.columns:
    le = LabelEncoder()
    df1_encode[column] = le.fit_transform(df1_encode[column])
    label_encoders[column] = le

# print each label related content
for column, le in label_encoders.items():
    print(f"Column: {column}")
    for class_index, class_label in enumerate(le.classes_):
        print(f"  {class_index}: {class_label}")

# print df1_encode
print(df1_encode)
```

```python
rename = {'model': 'model1', 'storage_int': 'storage1', 'colour': 'colour1'}
df1_encode = df1_encode.rename(columns=rename)
df1_encode
```

```python
df1=pd.concat([df1,df1_encode],axis=1)
df1
```

```python
df1['product_id'] = df1['model1'].astype(str) + df1['storage1'].astype(str) + df1['colour1'].astype(str)
```

```python
df1['merge_key_2'] = df1['model1'].astype(str) + df1['storage1'].astype(str)
```

```python
df1
```

```python
df1= df1[['product_id', 'product_name', 'product_price','product_rating','model', 'storage_int', 'colour','merge_key_2']]
df1
```

```python
#data type conversion
#remove '$'from product_price
#remove ',' from price
df3['product_price'] = df3['product_price'].str.replace('$', '', regex=False).str.replace(',', '').astype(float)
df3
```

```python
#divide column product_name
df3=pd.DataFrame(df3)
```

```python
# identify the pattern of column 'product_name' for spliting
pattern3 = r'(?P<brand>\w+)\s(?P<model>(?:iPhone\s)?[\w\s]+?)\s(?:5G\s)?(?P<storage>\d+GB)\s\s(?P<colour>[\w\s]+)'

# split column
df3[['brand', 'model', 'storage', 'colour']] = df3['product_name'].str.extract(pattern3)

print(df3)
```

```python
df3.loc[:2, 'model'] = df3.loc[:2, 'model'].str.replace(' 5G 6GB RAM', '')
```

```python
df3_encode = df3[['model', 'storage', 'colour']]
print(df3_encode)
```

```python
# convert_storage_to_int for df3_encode
def convert_storage_to_int(storage):
    return int(storage.replace('GB', '').replace('TB', '000')) if 'TB' in storage else int(storage.replace('GB', ''))
```

```python
df3_encode['storage'] = df3_encode['storage'].apply(convert_storage_to_int)
```

```python
df3_encode
```

```python
#convert data type of 'storage' to string
df3_encode['storage'] = df3_encode['storage'].astype(str)
```

```python
# define mapping dictionary bases on df1 encoded record
model_mapping = {
    'iPhone 11': 0,
    'iPhone 12': 1,
    'iPhone 13': 2,
    'iPhone 14': 3,
    'iPhone 15': 4,
    'iPhone 15 Plus': 5,
    'iPhone 15 Pro': 6,
    'iPhone 15 Pro Max': 7
}

storage_mapping = {
    '64': 0,
    '128': 1,
    '256': 2,
    '512': 3,
    '1000': 4
}

colour_mapping = {
    'Black': 0,
    'Black Titanium': 1,
    'Blue': 2,
    'Blue Titanium': 3,
    'Green': 4,
    'Midnight': 5,
    'Natural Titanium': 6,
    'Pink': 7,
    'Starlight': 8,
    'White': 9,
    'White Titanium': 10,
    'Yellow': 11
}

# encode columns bases on mapping dictionary
df3_encode['model3'] = df3_encode['model'].map(model_mapping)
df3_encode['storage3'] = df3_encode['storage'].map(storage_mapping)
df3_encode['colour3'] = df3_encode['colour'].map(colour_mapping)

# print df3_encode
print(df3_encode)
```

```python
df3_encode=df3_encode[['model3','storage3','colour3']]
df3_encode
```

```python
df3=pd.concat([df3,df3_encode],axis=1)
df3
```

```python
df3['product_id'] = df3['model3'].astype(str) + df3['storage3'].astype(str) + df3['colour3'].astype(str)
```
