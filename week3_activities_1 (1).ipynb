```python
#Text Data Cleaning:

df.rename(columns={'Unnamed: 0': 'Country', '? Summer': 'Summer', '01 !': 'Gold_s','02 !':'Silver_s','03 !':'Bronze_s','Total':'Total_summer','? Winter':'Winter','01 !.1':'Gold_w','02 !.1':'Silver_w','03 !.1':'Bronze_w','Total.1':'Total_Winter','? Games':'Total_games','01 !.2':'Tot_Gold','02 !.2':'Tol_Sil','03 !.2':'Tol_Bro','Combined total':'Combined_total'}, inplace=True)

df.info()
```

```python
# exclude the last row from df since it is subtotal row
df_new = df.iloc[:-1]
df_new.info()
```

```python
#check missing value
missing_values=df_new.isnull().sum()
print(missing_values)
```

```python
#Duplicate value?
format(len(df_new[df_new.duplicated()]))
```

```python
#check datatype(olympics)
df_new.info()
```

```python
#check structual error
df_new.loc[~df_new['Country'].astype(str).str.isdigit(),'Country'].tolist()
#[1.15,'Armenia']
```

```python
#Check for Data Integrity:
# split the first column "country" to three new columns base on its country name, county code, and country code 2
df_new[['Country Name', 'Country Code', 'Country Code 2']] = df['Country'].str.extract(r'([\w\s]+) \((\w+)\)(?: \[(\w+)\])?')

# check the adjusted dataframe
print(df_new.head())
```

```python
# drop the first column 'Country'
df_new1=df_new.drop(columns=['Country'])
df_new1.info()
```

```python
# Move 'Country Name'、'Country Code' 和 'Country Code 2' to the front
df_new2 = df_new1[['Country Name', 'Country Code', 'Country Code 2'] + [col for col in df_new1.columns if col not in ['Country Name', 'Country Code', 'Country Code 2']]]

# check the adjusted dataframe
print(df_new2.head())

```

```python
df_new2.info()
```

```python
# According to the  'Country Code' value to fulfil 'Country Code 2' missing value
df_new2['Country Code 2'] = df_new2['Country Code 2'].fillna(df_new2['Country Code'])
df_new2.info()
```

```python
#check structual error
df_new2.loc[~df_new['Country Code 2'].astype(str).str.isdigit(),'Country Code 2'].tolist()
#[1.15,'Armenia']
```

```python
string_columns = ['Country Name', 'Country Code', 'Country Code 2']

# save the string dataframe
string_df = df_new2[string_columns]
print(string_df)
```

```python
#find outliner
import matplotlib.pyplot as plt

numerical_columns = ['Summer','Gold_s','Silver_s', 'Bronze_s', 'Total_summer','Winter','Gold_w','Silver_w','Bronze_w','Total_Winter','Total_games','Tot_Gold','Tol_Sil','Tol_Bro','Combined_total']

# find the mean
mean_values = df_new2[numerical_columns].mean()

# print the mean
print(mean_values)
```

```python
from scipy.stats import zscore

# select zscore caculation columns(except 1st to 3rd columns)
numerical_columns = df_new2.columns[3:]

# caculation
z_scores = df_new2[numerical_columns].apply(zscore)

threshold = 3

# identify outliner
outliers = np.abs(z_scores) > threshold

# using mean to replace outliner
df_new2.iloc[1:, 1:] = df_new2.iloc[1:, 1:].where(~outliers, np.nanmean(df_new2[numerical_columns]), axis=0)
```

```python
# find the mean
mean_values = df_new2[numerical_columns].mean()

# print the mean and compare to previous mean, value decreased.
print(mean_values)
```

```python
# normalise dataframe
from sklearn.preprocessing import MinMaxScaler

# select numberical columns
numerical_columns = df_new2.select_dtypes(include=['int64', 'float64']).columns
numerical_df = df_new2[numerical_columns]

# initiate MinMaxScaler
scaler = MinMaxScaler()

# normalise numberical data
normalized_df = scaler.fit_transform(numerical_df)

# conver the normalise data to dataframe
normalized_df = pd.DataFrame(normalized_df, columns=numerical_columns)

# print the normalized DataFrame
print(normalized_df.head())

```

```python
# using histogram to visualise cleaned date
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

for i, column in enumerate(normalized_df.columns):
    plt.subplot(4, 4, i + 1)
    plt.hist(normalized_df[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(column)
    plt.xlabel('Normalized Value')
    plt.ylabel('Frequency')

plt.tight_layout()

plt.show()
```

```python
numerical_columns = df_new2.columns[3:]
plt.figure(figsize=(6, 4))

# Iterate over each numerical column and plot histogram
for column in numerical_columns:
    plt.hist(df_new2[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(column)
    plt.xlabel('Original Value')
    plt.ylabel('Frequency')
    plt.show()  # Show each histogram separately

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Combine the string part and the normalized part
final_df = pd.concat([string_df, normalized_df], axis=1)
print(final_df)
```

```python
# save as csv
final_df.to_csv("/content/drive/MyDrive/Data Transformation/combined_data.csv", index=False)
```
