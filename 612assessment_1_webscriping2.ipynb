```python
import requests
from bs4 import BeautifulSoup
import csv

```

```python
# URL of the webpage to scrape
url = "https://reebelo.co.nz/collections/apple-iphone?brand=Apple&sort=latest-release"

# Send a GET request to the URL
response = requests.get(url)

response
```

```python
# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# create an empty list to store the useful product information
all_items = []

# locate the useful information area
products = soup.select('.flex.flex-col.border.border-gray-100.bg-white')

# extract product name, price
for product in products:
    # product name
    name = product.select_one('h4').text.strip()

    # product price
    price = product.select_one('div.mt-1.text-center.text-xs span').text.strip()

    # store each product inforamtion into dictionary
    item = {
        'product_name': name,
        'product_price': price
    }

    # add dictionary into the list.
    all_items.append(item)

# conver the list into csv format
with open('all_products.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=['product_name', 'product_price'])
    writer.writeheader()
    writer.writerows(all_items)

print("All information has been saved to all_products.csv")
```
