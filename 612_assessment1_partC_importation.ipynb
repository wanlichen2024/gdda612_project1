```python
import pandas as pd
import numpy as np
```

```python
#read df1, df2,df3
df1=pd.read_csv('/content/drive/MyDrive/iPhone_noelleeming.csv')
df2=pd.read_csv('/content/drive/MyDrive/iphone_reebelo.csv')
df3=pd.read_csv('/content/drive/MyDrive/iphone_phonewarehouse.csv')
```

```python
# convert to DataFrame
df1=pd.DataFrame(df1)
df2=pd.DataFrame(df2)
df3=pd.DataFrame(df3)
```

```python
merged_df = pd.merge(df1, df3, on='product_id', how='left')
```

```python
final_df = pd.merge(merged_df, df2, on='merge_key_2', how='left')
```

```python
final_df.shape
```

```python
final_df= final_df.rename(columns={'product_price_x': 'Noel_leeming_price'})
```

```python
final_df= final_df.rename(columns={'product_price_y': 'Phone_warehouse_price'})
```

```python
final_df= final_df.rename(columns={'product_price': 'Reebelo_price'})
```

```python
# keep useful columns for final_df
final_df=final_df[['product_id','product_name_x','storage_int','colour_x','product_rating','Noel_leeming_price','Phone_warehouse_price','Reebelo_price']]
```

```python
#rename columns
final_df=final_df.rename(columns={'product_name_x': 'product_name'})
final_df=final_df.rename(columns={'colour_x': 'colour'})
final_df=final_df.rename(columns={'storage_int': 'storage_G'})
```

```python
# a. set a specific column as the index
index_product_id = final_df.set_index('product_id')
print("\n df with 'product_id' as index:\n", index_product_id)
```

```python
# b. reset to default integer index
index_reset = final_df.reset_index()
print("\n reset to default integer-based index:\n", index_reset)
```

```python
#c. selecting rows by conditional index
cond = index_reset['product_rating'] > 4.0 #condition is rating over 4.0
conditional_df = index_reset[cond]
print("\n Rows' rating > 4.0:\n", conditional_df)
```

```python
# d. multi-level indexing
multi_index = index_reset.set_index(['product_rating', 'storage_G'])
print("\n multi-level index ('product_rating' and 'storage_x'):\n", multi_index)
```

```python
# Task 8: Sorting

# a. sort by product rating , ascending
sorted_asc = final_df.sort_values(by='product_rating', ascending=True)
print("\n sorted by product rating, ascending:\n", sorted_asc)
```

```python
# b. sort by product rating, descending
sorted_des = final_df.sort_values(by='product_rating', ascending=False)
print("\n sorted by product rating, descending:\n", sorted_des)
```

```python
# Task 9: Summary Statistics
sum_stats = final_df.describe()
print("Summary Statistics:\n", sum_stats)
```
