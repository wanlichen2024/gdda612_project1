```python
np.random.seed(42)

total_elements = df.size

# 计算需要设置为缺失值的元素数量
num_missing = int(total_elements * 0.001)

# 随机选择要设置为缺失值的元素的位置
missing_positions = np.random.choice(total_elements, num_missing, replace=False)

# 将数据框展平为一维数组，以便于索引
flattened_df = df.values.flatten()

# 将选定的位置设置为 NaN
flattened_df[missing_positions] = np.nan

# 将数据重新整形为原来的形状
df_with_missing = pd.DataFrame(flattened_df.reshape(df.shape), columns=df.columns)

# 查看结果
print(df_with_missing.isna().sum())
```

```python
df= df_with_missing
```

```python
# checking missing value
df.isnull().sum()
```

```python
# Deal with missing Value
# 1. Using Mean to fill missing value
df['Customer_Age'].fillna(df['Customer_Age'].mean(), inplace=True)
df['Dependent_count'].fillna(df['Dependent_count'].mean(), inplace=True)
df['Months_on_book'].fillna(df['Months_on_book'].mean(), inplace=True)
df['Credit_Limit'].fillna(df['Credit_Limit'].mean(), inplace=True)
df['Total_Relationship_Count'].fillna(df['Total_Relationship_Count'].mean(), inplace=True)
df['Months_Inactive_12_mon'].fillna(df['Months_Inactive_12_mon'].mean(), inplace=True)
df['Contacts_Count_12_mon'].fillna(df['Contacts_Count_12_mon'].mean(), inplace=True)
df['Total_Revolving_Bal'].fillna(df['Total_Revolving_Bal'].mean(), inplace=True)
df['Avg_Open_To_Buy'].fillna(df['Avg_Open_To_Buy'].mean(), inplace=True)
df['Total_Amt_Chng_Q4_Q1'].fillna(df['Total_Amt_Chng_Q4_Q1'].mean(), inplace=True)
df['Total_Trans_Amt'].fillna(df['Total_Trans_Amt'].mean(), inplace=True)
df['Total_Trans_Ct'].fillna(df['Total_Trans_Ct'].mean(), inplace=True)
df['Total_Ct_Chng_Q4_Q1'].fillna(df['Total_Ct_Chng_Q4_Q1'].mean(), inplace=True)
df['Avg_Utilization_Ratio'].fillna(df['Avg_Utilization_Ratio'].mean(), inplace=True)
```

```python
# 2.Since client number should be a unique identity number, dropping the missing value is the most suitable way.
# Attrition_Flag is an important colunm for later predition, so dropping the missing value is a proper way.
df.dropna(subset=['CLIENTNUM'],inplace=True)
df.dropna(subset=['Attrition_Flag'],inplace=True)
```

```python
# 3. Fulful catergory vairables as mode
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df['Education_Level'].fillna(df['Education_Level'].mode()[0], inplace=True)
df['Marital_Status'].fillna(df['Marital_Status'].mode()[0], inplace=True)
df['Card_Category'].fillna(df['Card_Category'].mode()[0], inplace=True)
```

```python
# 4. fill income_category as 'Unknown'
df['Income_Category'].fillna('Unknown', inplace=True)
```
